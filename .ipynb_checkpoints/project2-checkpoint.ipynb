```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, classification_report
```

# data

```python
import pandas as pd
df=pd.read_csv("healthcare-dataset-stroke-data.csv")

df
```

```python
df.info()
```

# data preproccesing

```python
df.isnull().sum()
df.duplicated()
```

```python
df=df.dropna()
```

```python
lb=LabelEncoder()
st=StandardScaler()
```

```python
df.iloc[:,1]=lb.fit_transform(df.iloc[:,1].values)
df.iloc[:,5]=lb.fit_transform(df.iloc[:,5].values)
df.iloc[:,6]=lb.fit_transform(df.iloc[:,6].values)
df.iloc[:,7]=lb.fit_transform(df.iloc[:,7].values)
df.iloc[:,10]=lb.fit_transform(df.iloc[:,10].values)

df
```

```python
df.corr()
```

# heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.iloc[:,1:12].corr(), annot=True)
```

```python
x=df.drop(['stroke','id'],axis=1)
y=df['stroke']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# model

```python
#svc
SVM_rbf = SVC(kernel='rbf', random_state=100, gamma=3.0,C=1.5 )
svm_classifier=SVM_rbf.fit(x_train,y_train)
y_test_prediction_svm = svm_classifier.predict(x_test)
print("SVM:\n",y_test_prediction_svm)
print("The confusion Matrix:\n",confusion_matrix(y_test, y_test_prediction_svm))

#nb
nbc=GaussianNB()
nbc1=nbc.fit(x_train,y_train)
y_test_prediction_nbc=nbc1.predict(x_test)
print("GaussianNB:\n",y_test_prediction_nbc)
print("the confusion matrix:\n",confusion_matrix(y_test, y_test_prediction_nbc))
#dt
dt=DecisionTreeClassifier()
dt1=dt.fit(x_train,y_train)
y_test_prediction=dt1.predict(x_test)
print("dt:\n",y_test_prediction)
print("confusion matrix:\n",confusion_matrix(y_test,y_test_prediction))

#LR
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
y_test_prediction_lr=logreg.predict(x_test)
print("LR:\n",y_test_prediction_lr)
print("The confusion Matrix:\n",confusion_matrix(y_test, y_test_prediction_lr))

#KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_test_prediction_knn = knn.predict(x_test)
print("KNN:\n",y_test_prediction_knn)
print("The confusion Matrix:\n",confusion_matrix(y_test, y_test_prediction_knn))
```

# Accuracy

# SVM Accuracy 95%

```python
accuracy_score(y_test,svm_classifier.predict(x_test))
```

```python
print(classification_report(y_test,svm_classifier.predict(x_test)))
```
