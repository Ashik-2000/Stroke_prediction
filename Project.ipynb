```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
import seaborn as sb
```

# Dataset Import

```python
data = pd.read_csv('lcancer.csv')
```

```python
data.info()
```

# Data Preprocessing

```python
data = data.drop(columns=['index','Patient Id'])
data['Level'] = preprocessing.LabelEncoder().fit_transform(data['Level'])
```

```python
x = data.drop(columns = ['Level'])
y = data['Level']
# High = 0
# Low = 1
# Medium = 2
```

# Data Train & Testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.20)
```

# ML Algorithms Apply

```python
model1 = DecisionTreeClassifier()
model1.fit(x_train, y_train)
predict1 = model1.predict(x_test)
cf = confusion_matrix(y_test,predict1)
accuracy = accuracy_score(y_test, predict1)
print('The confusion matrix is:\n',cf)
print('The accuracy score is: ',accuracy)
print("\nThe classification report:\n",classification_report(y_test,predict1))
```

```python
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
predict2 = model2.predict(x_test)
cf2 = confusion_matrix(y_test,predict2)
accuracy2 = accuracy_score(y_test, predict2)
print('The confusion matrix is:\n',cf2)
print('The accuracy score is: ',accuracy2)
print("\nThe classification report:\n",classification_report(y_test,predict2))
```

```python
model3 = BernoulliNB()
model3.fit(x_train, y_train)
predict3 = model3.predict(x_test)
cf3 = confusion_matrix(y_test,predict3)
accuracy3 = accuracy_score(y_test, predict3)
print('The confusion matrix is:\n',cf3)
print('The accuracy score is: ',accuracy3)
print("\nThe classification report:\n",classification_report(y_test,predict3))
```

```python
model4 = KNeighborsClassifier()
model4.fit(x_train, y_train)
predict4 = model4.predict(x_test)
cf4 = confusion_matrix(y_test,predict4)
accuracy4 = accuracy_score(y_test, predict4)
print('The confusion matrix is:\n',cf4)
print('The accuracy score is: ',accuracy4)
print("\nThe classification report:\n",classification_report(y_test,predict4))
```

```python
model5 = SVC()
model5.fit(x_train, y_train)
predict5 = model5.predict(x_test)
cf5 = confusion_matrix(y_test,predict5)
accuracy5 = accuracy_score(y_test, predict5)
print('The confusion matrix is:\n',cf5)
print('The accuracy score is: ',accuracy5)
print("\nThe classification report:\n",classification_report(y_test,predict5))
```

```python
model6 = MLPClassifier()
model6.fit(x_train, y_train)
predict6 = model6.predict(x_test)
cf6 = confusion_matrix(y_test,predict6)
accuracy6 = accuracy_score(y_test, predict6)
print('The confusion matrix is:\n',cf6)
print('The accuracy score is: ',accuracy6)
print("\nThe classification report:\n",classification_report(y_test,predict6))
```
